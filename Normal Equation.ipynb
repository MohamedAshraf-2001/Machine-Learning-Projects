```python
import numpy as np 
import matplotlib.pyplot as plt 
```

## Normal Equation
1) **fast** with large n (no. instances).
2) **slow** with large m (no. features ).
3) **zero** Hyperparameters.
3) **No** scaling required.

# Code from Scratch

```python
x = 2 * np.random.randn(100,1)   #x1 ==> single feature, x0=1
y = 4 + 3*x + np.random.randn(100,1)  #y = 4 + 3*x1 + Gaussian noise
                                      #theta_0 = 4, theta_1 = 3
plt.scatter(x,y)
plt.xlabel('feature')
plt.ylabel('target')
plt.title('random dataset')
```

```python
x_b = np.c_[np.ones((100, 1)), x]  #add x0 to each instance x1
theta_target = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
theta_target
```

```python
theta_0_new = theta_target[0]
theta_1_new = theta_target[1]
print(theta_0_new,theta_1_new)
```

```python
x_new = np.array([[0],[2]])
x_new_b = np.c_[np.ones((2,1)), x_new]
x_new_b
```

```python
y_pred = x_new_b.dot(theta_target)
y_pred
```

```python
plt.plot(x_new, y_pred, "r-")
plt.scatter(x, y)
plt.axis([0, 2, 0, 15])
```
