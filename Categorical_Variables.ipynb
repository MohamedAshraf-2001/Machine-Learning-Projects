```python
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Datasets/melb_data.csv')
df.head()
```

```python
df.info()
```

```python
X = df.drop(['Price'], axis = 1)
y = df.Price
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, 
                                                      test_size=0.2, random_state=0)
```

```python
#drop columns with missing values (simplest approach)
cols_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()]
cols_with_missing
```

```python
X_train.drop(cols_with_missing, axis = 1, inplace = True)
X_valid.drop(cols_with_missing, axis = 1, inplace = True)
```

```python
df.nunique()
```

```python
# "Cardinality" means the number of unique values in a column
# Select categorical columns with relatively low cardinality (convenient but arbitrary)
low_cardinality_cols = [cname for cname in X_train.columns if X_train[cname].nunique() < 10 and X_train[cname].dtype == 'object']
```

```python
low_cardinality_cols
```

```python
numerical_cols = [cname for cname in X_train.columns if X_train[cname].dtype in ['int64', 'float64']]
```

```python
numerical_cols
```

```python
my_cols = low_cardinality_cols + numerical_cols
X_train = X_train[my_cols].copy()
X_valid = X_valid[my_cols].copy()
```

```python
X_train
```

```python
#get list of categorical variables 
s = (X_train.dtypes == 'object')
print(s)
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_test):
    model = RandomForestRegressor(n_estimators = 100, random_state = 0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
#One hot-encoding
from sklearn.preprocessing import OneHotEncoder

OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_train = pd.DataFrame(OH_encoder.fit_transform(X_train[object_cols]))
OH_valid = pd.DataFrame(OH_encoder.transform(X_valid[object_cols]))

#one hot-encoding removes index
OH_train.index = X_train.index
OH_valid.index = X_valid.index

#remove categorical columns (will replace one-hot encoding)
num_X_train = X_train.drop(object_cols, axis = 1)
num_X_valid = X_valid.drop(object_cols, axis = 1)

OH_X_train = pd.concat([num_X_train, OH_train], axis = 1)
OH_X_valid = pd.concat([num_X_valid, OH_valid], axis = 1)
```

```python
print("MAE from (One-Hot Encoding):") 
print(score_dataset(OH_X_train, OH_X_valid, y_train, y_valid))
```
