```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu')
accuracy_score(y_test, y_pred)
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = df.iloc[:, 0:2].values
y = df.iloc[:, -1].values
```

```python
X_new_age = X[:,0] * X[:,0]
X_new_salary = X[:,1] * X[:,1]
```

```python
X_new_age = pd.Series(X_new_age)
X_new_age
```

```python
X_new_salary = pd.Series(X_new_salary)
X_new_salary
```

```python
df=df.drop(['Purchased'], axis=1)
df
```

```python
df = df.assign(new_age=X_new_age)
df = df.assign(new_salary=X_new_salary)
df = df.assign(Purchased = y)
df
```

```python
X1 = df.iloc[:, 0].values 
X2  = df.iloc[:, 1].values
X3 = df.iloc[:, 2].values 
X4 = df.iloc[:, 3].values 
```

```python
X = df.iloc[:, 0:4].values
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X, y,test_size = 0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver = 'liblinear', random_state = 0)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu')
accuracy_score(y_test, y_pred)
```
